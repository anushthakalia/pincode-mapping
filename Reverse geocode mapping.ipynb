```python
import numpy as np
from scipy import spatial
import pandas as pd
```

```python
postaldata = pd.read_csv('AllIndiaPostalCodesone.csv')
postaldata
```

```python
latlongs = postaldata.iloc[:,2:]
latlongs
```

```python
data =  np.array(latlongs)
```

```python
tree = spatial.KDTree(data)
```

```python
latlongs = np.array([18.480326,73.858055])
result = tree.query(latlongs)
result
```
